# Clean Up Data

Start by importing dependencies.

```python
import pandas as pd
import numpy as np
```

Import the 2025 dataset. This notebook should be run from the root project directory so Pandas can find the csv data file.

```python
wildfires_dirty = pd.read_csv('../data/wildfires_2025.csv')
print("Initial dataset shape: ", wildfires_dirty.shape)
```

Drop invalid rows (containing `32767`) and an unnamed column. Note that 32,767 is the maximum value of a two's complement 16-bit integer.

```python
# Identity invalid rows (rows with the value of 32,767 in any column)
valid_rows = wildfires_dirty.agg(lambda rows: np.all(rows != 32767), axis="columns")
print(f"Dropping {np.logical_not(valid_rows).sum()} invalid rows")

# Drop invalid rows and the "unnamed" column.
wildfires_clean = wildfires_dirty[valid_rows] \
    .drop(columns="Unnamed: 0") \
    .reset_index(drop=True) 
```

Preview final result.

```python
print("Final shape:", wildfires_clean.shape)
wildfires_clean
```

Export cleaned-up CSV file.

```python
wildfires_clean.to_csv("../data/wildfires_clean2025_py.csv", index=None)
```
